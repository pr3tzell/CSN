Scrum: Como un equipo se organiza para no perderse

Sprint: Ciclos cortos de trabajo
- Sirve para no perderse en equipo

Sin Scrum:
- Nadie sabe quien hace que
- Todos hablan al mismo tiempo
- Al final nadie funciona

Con Scrum:
- Sabemos quien hace que

Que es Scrum?

Elegir: Decidir que hacemos cada semana

Hacer: Trabajamos juntos en lo elegido

Mostrar: Enseñamos lo que quedo

3 roles:
- Po (Product Owner): Dice que hacer
- Sm(Scrum Master): Cuida el equipo y resuelve problemas
- EQ(Equipo): Decide como hacer las cosas y las construye

La daily en 3 preguntas
- Que hice ayer?
- Que hare hoy?
- Que me bloquea?

El backlog: Todo lo que hay que hacer
1. Elegir destino
2. Pedir permiso al colegio
3.Cotizar bus
4.Cobrar el dinero

Errores tipicos al hacer scrum:
- Si la reunion una hora, no es daily
- El SM no es jefe, solo ayuda al equipo a avanzar
- Meter tareas nuevas rompe el sprint
- Sin mirar atras, el equipo nunca avanza

1. Sprint: Un ciclo corto donde el equipo entrega algo

```python
class TarjetaCredito:
  def __init__(self, banco, limite):
    self.banco = banco
    self.limite = limite

 def mostrar_limite(self):
  print(f"tarjeta de {self.banco} con limite de ${self.limite}")

class UsuarioConTarjeta:
  def __init__(self, nombre, tarjeta):
    self.nombre = nombre
    self.tarjeta = tarjeta

  def mostrar_tarjeta(self):
    print(f"Usuario: {self.nombre}")
    self.tarjeta.mostrar_limite()

mi_tarjeta = TarjetaCredito("Banco Demo", 1000000)
usuario = UsuarioConTarjeta("Ana", mi_tarjeta)
usuario.mostrar_tarjeta()
```
